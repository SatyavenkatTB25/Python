1Q) Write a Python program to generate the following plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Create arrays for x and y coordinates
x = np.array([0, 10])
y = np.array([0, 10])

# Create a figure and axis (subplot)
fig, axs = plt.subplots()

# Set the x and y axis limits
axs.set_xlim(0, 10)
axs.set_ylim(0, 10)

# Plot the data using a blue line with a label
plt.plot(x, y, label='linear', color='blue', linewidth=1)

# Add a legend to the plot
axs.legend(loc='upper right')

# Display the plot
plt.show()

```

2Q) Plot a pie chart using the following information using appropriate labels and different colors

```python
import matplotlib.pyplot as plt

# Data
courses = ['I501', 'B530', 'B535', 'B581', 'B583']
no_of_students = [72, 49, 40, 42, 23]

# Define custom colors for the pie chart
colors = ['blue', 'green', 'yellow', 'orange','red']

# Create a pie chart
plt.pie(no_of_students, colors=colors, autopct='%1.1f%%')

# Set the title for the pie chart
plt.title('Distribution of Students according to the Course')

# Add a legend with course names
plt.legend(courses, title="Courses", loc="upper right", bbox_to_anchor=(1, 0, 0.1, 1))

# Set the aspect ratio to make the pie chart a circle
plt.axis('equal')

# Display the pie chart
plt.show()
```

3Q) Write a python program which has the following data [1,2,3][3,4,5] and plots a vertical bar chart, horizontal bar chart and a blank plot in one pyplot figure, as shown below

```python
import matplotlib.pyplot as plt

# Create a figure
fig = plt.figure()

# Add three subplots
f1 = fig.add_subplot(131)  # First subplot
f2 = fig.add_subplot(132)  # Second subplot
f3 = fig.add_subplot(133)  # Third subplot

# Data
x = [1, 2, 3]
y = [3, 4, 5]

# Set x and y limits for the first subplot
f1.set_xlim(1, 4)
f1.set_ylim(0, 5)

# Set x and y limits for the second subplot
f2.set_xlim(0, 5)
f2.set_ylim(1, 4)

# Set x and y limits for the third subplot
f3.set_xlim(0, 1)
f3.set_ylim(0, 1)

# Set major tick locators for the x-axes of the subplots
f1.xaxis.set_major_locator(plt.MultipleLocator(0.5))
f2.xaxis.set_major_locator(plt.MultipleLocator(1))
f3.xaxis.set_major_locator(plt.MultipleLocator(0.2))

# Set tick parameters for all subplots to display ticks on top and right sides
f1.tick_params(top=True, right=True, direction="in")
f2.tick_params(top=True, right=True, direction="in")
f3.tick_params(top=True, right=True, direction='in')

# Create a bar plot in the first subplot
f1.bar(x, y, color='b', align='edge', edgecolor='k')

# Create a horizontal bar plot in the second subplot
f2.barh(x, y, color='b', align='edge', edgecolor='k')

# Display the plots
plt.show()
```

4Q) Create a pie chart like the one below, with labels: A, B, C, D, E; and the numbers associated with each label is: 3, 12, 16, 2, 5 respectively. Give each part a different color (not required to be same as shown below), a title Question 4, list labels on the chart.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
numbers = np.array([3, 12, 16, 2, 5])
labels = ['A', 'B', 'C', 'D', 'E']
colors = ['blue', 'green', 'red', 'cyan', 'magenta']

# Create a pie chart using plt.pie() with specified labels, colors, and edgecolor
plt.pie(numbers, labels=labels, colors=colors, wedgeprops={"edgecolor": "k"})

# Set axis scaling to "auto" to ensure the aspect ratio is maintained
plt.axis("auto")

# Set the title for the pie chart
plt.title('Question 4')

# Display the pie chart
plt.show()
```

5Q) Plot a graph which compares the simple interest and the compound interest of the principal amount of 5000 for 10 years at the interest rate of 15%. Please create two line charts in a single plot with appropriate labels and different colors/styles

```python
import matplotlib.pyplot as plt

# Constants
principal = 5000
rate = 0.15
years = 10

# Calculate simple interest and compound interest for each year
simple_interest = [principal * rate * year for year in range(1, years + 1)]

compound_interest = []
current_principal = principal
for year in range(1, years + 1):
    interest = current_principal * rate
    current_principal += interest
    compound_interest.append(current_principal - principal)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot simple interest in blue
plt.plot(range(1, years + 1), simple_interest, label='Simple Interest', color='blue', linestyle='-', marker='o')

# Plot compound interest in red
plt.plot(range(1, years + 1), compound_interest, label='Compound Interest', color='red', linestyle='--', marker='x')

# Add labels and legend
plt.title('Comparison of Simple Interest and Compound Interest')
plt.xlabel('Years')
plt.ylabel('Interest Amount')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

6Q) For the two matrices below, please write a Python program using numpy arrays that calculates their sum and dot product. A= [[1 0 0]  B= [[1 1 1]]
                               [0 2 0]      [1 1 1]
                               [0 0 3]]     [1 1 1]]

```python
import numpy as np

# Create matrices A and B
A = np.array([[1, 0, 0], [0, 2, 0], [0, 0, 3]])
B = np.array([[1, 1, 1], [1, 1, 1], [1, 1, 1]])

# To Calculate the sum of matrices A and B
matrix_sum = A + B

# To Calculate the dot product of matrices A and B
matrix_dot_product = np.dot(A, B)

# Create a random matrix with values between 1 and 9 (inclusive)
random_matrix = np.random.randint(1, 10, size=(3, 3))

# Calculate the total sum of matrices A, B, and the random matrix
total_sum = matrix_sum + random_matrix

# Print Matrix A
print("Matrix A:")
print(A)

# Print Matrix B
print("\nMatrix B:")
print(B)

# Print the sum of Matrix A and Matrix B
print("\nSum of Matrix A and Matrix B is:")
print(matrix_sum)

# Print the dot product of Matrix A and Matrix B
print("\nDot Product of Matrix A and Matrix B is:")
print(matrix_dot_product)

# Print the randomly generated matrix
print("\nRandomly Generated Matrix:")
print(random_matrix)

# Print the total sum (Matrix A + Matrix B + Random Matrix)
print("\nTotal Sum (Matrix A + Matrix B + Random Matrix):")
print(total_sum)
```
