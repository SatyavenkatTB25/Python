1Q) The scores from a class test are as follows: five students scored 12, three students scored 60, four students scored 72, six students scored 100 points. Write a Python program to print the Mean, Median, Variance and Standard deviation of the scores

```python
import statistics

# List of scores
scores = [12, 12, 12, 12, 12, 60, 60, 60, 72, 72, 72, 72, 100, 100, 100, 100, 100, 100]

# Calculate the mean
mean = sum(scores) / len(scores)

# Calculate the median
median = statistics.median(scores)

# Calculate the variance
variance = statistics.variance(scores)

# Calculate the standard deviation
std_dev = statistics.stdev(scores)

# Print the results
print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

2Q) The following table represents the frequency of the weights of 25 students in a team: -
weights = [88, 76, 65, 96, 52, 98, 102]
frequency = [4, 3, 5, 2, 5, 2, 4]
Write a python program to create a histogram (using pyplot) for the above table.

```python
import matplotlib.pyplot as plt

# Data from the frequency table
weights = [88, 76, 65, 96, 52, 98, 102]
frequency = [4, 3, 5, 2, 5, 2, 4]

# Create a list of weights based on the frequency
data = []
for w, f in zip(weights, frequency):
    data.extend([w] * f)

# Create the histogram
plt.hist(data, bins=20, edgecolor='black')

# Add labels and title
plt.xlabel('Weight (lbs)')
plt.ylabel('Number of Students')
plt.title('Histogram of Student Weights')

# Show the histogram
plt.show()
```

3Q) For the data given in question 2, write a Python program to create a dictionary, which maps the values to their frequencies and calculates the probability of each weight using their frequencies.

```python
# Data from the frequency table
weights = [88, 76, 65, 96, 52, 98, 102]
frequency = [4, 3, 5, 2, 5, 2, 4]

# Create a dictionary to map values to their frequencies
weight_frequency_dict = dict(zip(weights, frequency))

# Calculate the total number of students
total_students = sum(frequency)

# Calculate the probability of each weight
probability_dict = {}
for weight, freq in weight_frequency_dict.items():
    probability = freq / total_students
    probability_dict[weight] = probability

# Print the dictionary mapping values to frequencies
print("Dictionary of Values to Frequencies:")
print(weight_frequency_dict)

# Print the probability of each weight
print("\nProbability of Each Weight:")
for weight, probability in probability_dict.items():
    print(f"Weight {weight} lbs: Probability {probability:.2f}")
```

4Q) Write a stochastic program in Python to replicate a fair die (one dice) roll, such that this one die is rolled 100 times. Display the output of the dice rolls as a histogram. 

```python
import random
import matplotlib.pyplot as plt

# Number of dice rolls
num_rolls = 100

# Simulate rolling the die 100 times and store the results
dice_rolls = [random.randint(1, 6) for _ in range(num_rolls)]

# Create the histogram
plt.hist(dice_rolls, bins=6, range=(1, 7), edgecolor='black', align='left')

# Add labels and title
plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.title('Histogram of 100 Die Rolls')

# Set the x-axis labels to the faces of the die
plt.xticks(range(1, 7))

# Show the histogram
plt.show()
```
